# Missing data

## 1. Loading data

```python
df = pd.read_csv(path_processed+'data_clean.csv')
df.info()
```

## 2. Visualizing missing data

```python
df_missingdata = df.columns[df.isna().any()].tolist()
```

### 2.1 Matrix visualization

```python
msno.matrix(df[df_missingdata])
```

### 2.2 Barchart visualization

```python
msno.bar(df[df_missingdata], color="blue", log=True, figsize=(30,18))
```

### 2.3 Heatmap visualization

```python
msno.heatmap(df[df_missingdata], figsize=(20,20))
```

### 2.4 Dendogram visualization

```python
msno.dendrogram(df[df_missingdata])
```
